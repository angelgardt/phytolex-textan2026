```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import networkx as nx

from collections import defaultdict
```

```python
dir = os.curdir + '/data'
files = os.listdir(dir)
```

```python
files
```

```python
crosscols = set(('id', 'name', 'lft', 'short_name', 'alt_label', 'level', 'comment', 'author', 'editor', 'creation_date_end', 'pages_in_publication', 'checked_by_linguist', 'notation', 'note', 'other_comment', 'encoding', 'definition', 'exact_match', 'rght', 'publication_volume', 'source', 'related', 'creation_date_start', 'language_comment', 'parent_id', 'tree_id'))
take_files = set(('citation',
 'copyoforiginal',
 'disease',
 'etymology',
 'function',
 'genre',
 'language',
 'lexeme',
 'lexeme_etymology',
 'lexeme_language_lexeme',
 'lexeme_parent_lexeme',
 'original',
 'original_genre',
 'original_language',
 'original_other_language',
 'plant',
 'plantpart',
 'plantusage',
 'plantusage_disease',
 'plantusage_disease_as_in_source',
 'plantusage_function',
 'plantusage_function_in_source',
 'plantusage_medicinal_form',
 'plantusage_plant_part_used',
 'plantusage_plant_part_used_in_source',
 'plantusage_vet_disease_as_in_source',
 'publication_copy_of_original',
 'publication_parent_publication',
 'usage',
 'usage_disease',
 'usage_function',
 'usage_language_affiliation',
 'usage_language_for_other_names',
 'usage_plant_part',
 'usage_plant_part_used',
 'usage_plant_usage',
 'usage_scientific_name'))
```

```python
def read_file(dir, f):
    fname = f[10:-4]
    if fname not in take_files:
        return None, None, None
    print(fname.upper())
    with open(dir + '/' + f, "r", encoding="utf-8") as file:
        df = pd.read_csv(file)
    for c in df.columns:
        if c in crosscols:
            df.rename(columns={c: f'{fname}_{c}'}, inplace=True)
    display(df.sample(3))
    return df, df.columns, fname
```

```python
dfs = {}   # dict of dataframes,  key = cut file name
cols = {}   # dict of dataframes cols as Index,  key = cut file name

for f in files:
    df, coll, name = read_file(dir, f)
    if name:
        dfs[name] = df
        cols[name] = coll
```

```python
cols_list = {}   # dict of dataframes cols as list,  key = cut file name
for f, col in cols.items():
    cols_list[f] = list(col)
    
cols_dict = defaultdict(list)
for f, col in cols_list.items():
    for c in col:
        cols_dict[c].append(f)
cols_dict = dict(sorted(cols_dict.items(), key=lambda x: (x[0], - len(x[1]))))   # sorted dict 'col name': files
len(cols_dict)
```

```python
crosscols_ = set(c for c, lf in cols_dict.items() if len(lf) > 1)   # set of common cols
len(crosscols_), len([c for c in crosscols_ if c[-3:] == '_id'])
```

```python
id_cols = set(c for c in cols_dict if c[-3:] == '_id')   # set of id cols for potential links
len(id_cols)
```

```python
def count_links(f, col1, col2):
    df = dfs[f][[col1, col2]].dropna(how='any')
    return int(df[col1].count())

col_graph = nx.Graph()

nlinks = []
for f, cs in cols_list.items():
    for i, c in enumerate(cs):
        if c in id_cols:
            for j in range(i + 1, len(cs)):
                if cs[j] in id_cols:
                    nlinks.append((c, cs[j], count_links(f, c, cs[j]), f, dfs[f].shape[0]))
                    col_graph.add_edge(c, cs[j], file=f, nlinks=nlinks[-1][2])
```

```python
plt.figure(figsize=(200, 200))
nx.draw_networkx(col_graph, with_labels=True, font_size=30)
plt.show();
```
